## 1) Importar librerias

Para este análisis vamos a usar `geopandas` y `matplotlib`

```python
# librerias requeridas
import geopandas as gpd
import matplotlib.pyplot as plt
```

## 2) Leer data

Los datos de los proyectos mineros se pueden descargar de aquí: https://datos.gob.ar/dataset/energia-proyectos-mineros-ubicacion-aproximada/archivo/energia_0c8317ac-eacb-4826-b7e7-8f0a15a62ac1 y un mapa base de Argentina se puede descargar de aquí: https://gadm.org/download_country_v3.html
  
Usando la función `gpd.read_file()`, primero leemos dos shapefiles: 
* **df** contiene información sobre los proyectos mineros
* **pais** contiene las fronteras de Argentina y límites interprovinciales

```python
# leer data
df = gpd.read_file("data/proyectos-mineros-ubicacin-aproximada-/proyectos-mineros-ubicacin-aproximada--shp.shp", encoding = 'utf-8')
pais = gpd.read_file("data/gadm36_ARG_shp/gadm36_ARG_1.shp", encoding = 'utf-8')
```

## 3) Explorar dataset

```python
# explorar dataset
print('\nInformación sobre el dataset:')
print(df.info())
```

Tenemos 77 observaciones y 6 variables. Para nuestros análisis usaremos las variables **ESTADO** y **MINERAL_PR**. A continuación, veremos la cantidad de observaciones según los valores de cada una de estas variables:

```python
print('\nCount de valores en variable ESTADO:')
print(df.ESTADO.value_counts())
print('\nCount de valores en variable MINERAL_PR:')
print(df.MINERAL_PR.value_counts())
```

## 4) Cambios en la data

En la variable **ESTADO** algunos valores tienen nombre muy largos, lo cual hará que no se vean bien en el gráfico de barras que haremos más adelante. Cambiaremos estos nombres largos por formas abreviadas:

```python
# cambiar algunos valores en la variable 'ESTADO'
df.loc[df.ESTADO == "Exploración avanzada", 'ESTADO'] = 'Exp avanzada'
df.loc[df.ESTADO == "Evaluación económica previa", 'ESTADO'] = 'Ev econ previa'
df.loc[df.ESTADO == "Proceso de Cierre", 'ESTADO'] = 'Proc cierre'
```

```python
# ver los cambios en los valores de la variable ESTADO
print('\nCount de valores en variable ESTADO:')
print(df.ESTADO.value_counts())
```

## 5) Gráficos de barra

Haremos dos gráficos de barra, que veremos lado a lado:
* Según el tipo de mineral (usando la variable 'MINERAL_PR')
* Según el estado del yacimiento (usando la variable 'ESTADO')

Para nuestros gráficos de barra necesitamos contar cuantas observaciones corresponden a cada valor de la variable.

```python
# group y summarize de los datos
nr_mineral = df.groupby('MINERAL_PR')['MINERAL_PR'].count().sort_values(ascending = False)
nr_estado = df.groupby('ESTADO')['ESTADO'].count().sort_values(ascending = False)
```

Habiendo creado las variables **nr_mineral** y **nr_estado**, podemos usarlas para graficar como sigue:

```python
# graficos de barra
# figure y axis
fig, ax = plt.subplots(1, 2, sharey = True)
plt.rcParams["figure.figsize"] = (15, 5)

# MINERAL_PR
nr_mineral.plot(kind = 'bar', rot = '45', ax = ax[0])
ax[0].set_xlabel("Mineral")
ax[0].set_ylabel("Frecuencia")
ax[0].set_title('Gráfico de barras por \n tipo de mineral')
ax[0].set_axisbelow(True)
ax[0].grid(color = 'gray', linestyle = 'dashed')

# ESTADO
nr_estado.plot(kind = 'bar', ax = ax[1])
ax[1].set_xlabel("Estado del yacimiento")
ax[1].set_title('Gráfico de barras por \n estado del yacimiento')
ax[1].set_axisbelow(True)
ax[1].grid(color = 'gray', linestyle = 'dashed')
plt.show()
```

## 6) Mapas

Ahora veremos cada uno de los yacimientos mineros sobre el mapa de Argentina. Esto es posible ya que nuestro **df** es un objeto del tipo `GeoDataFrame` y tiene una variable llamada **geometry** que contiene las coordenadas de cada observación.

```python
# subsets de los datos por tipo de mineral
carbon = df[df['MINERAL_PR'] == 'Carbón']
cobre = df[df['MINERAL_PR'] == 'Cobre']
hierro = df[df['MINERAL_PR'] == 'Hierro']
litio = df[df['MINERAL_PR'] == 'Litio']
oro = df[df['MINERAL_PR'] == 'Oro']
plata = df[df['MINERAL_PR'] == 'Plata']
plomo = df[df['MINERAL_PR'] == 'Plomo']
potasio = df[df['MINERAL_PR'] == 'Potasio']
uranio = df[df['MINERAL_PR'] == 'Uranio']
```

```python
# subsets de los datos por estado del yacimiento
exp_avanzada = df[df['ESTADO'] == 'Exp avanzada']
ev_econ_previa = df[df['ESTADO'] == 'Ev econ previa']
produccion = df[df['ESTADO'] == 'Producción']
factibilidad = df[df['ESTADO'] == 'Factibilidad']
prefactibilidad = df[df['ESTADO'] == 'Prefactibilidad']
mantenimiento = df[df['ESTADO'] == 'Mantenimiento']
construccion = df[df['ESTADO'] == 'Construcción']
proc_cierre = df[df['ESTADO'] == 'Proc cierre']
reingenieria = df[df['ESTADO'] == 'Reingeniería']
```

### 6.1) Mapa según tipo de mineral

```python
# mapa completo
# definir un mapa base con limites interprovinciales
ax = pais.plot(figsize = (10, 10), color = 'none', edgecolor = 'black')

# definir los nombres de los ejes y el titulo
ax.set_xlabel('Longitud')
ax.set_ylabel('Latitud')
ax.set_title('Proyectos mineros en Argentina')

# agregar los proyectos mineros segun tipo de mineral
carbon.plot(color = 'black', ax = ax, label = 'Carbón')
cobre.plot(color = 'brown', ax = ax, label = 'Cobre')
hierro.plot(color = 'red', ax = ax, label = 'Hierro')
litio.plot(color = 'blue', ax = ax, label = 'Litio')
oro.plot(color = 'gold', ax = ax, label = 'Oro')
plata.plot(color = 'grey', ax = ax, label = 'Plata')
plomo.plot(color = 'darkslategrey', ax = ax, label = 'Plomo')
potasio.plot(color = 'green', ax = ax, label = 'Potasio')
uranio.plot(color = 'lawngreen', ax = ax, label = 'Uranio')

# agregar leyenda
ax.legend(loc = 'lower right', title = 'Mineral')
```

### 6.2) Mapa según estado del yacimiento

```python
# definir un mapa base con limites interprovinciales
ax1 = pais.plot(figsize = (10, 10), color = 'none', edgecolor = 'black')

# definir los nombres de los ejes y el titulo
ax1.set_xlabel('Longitud')
ax1.set_ylabel('Latitud')
ax1.set_title('Proyectos mineros en Argentina')

# agregar los proyectos mineros segun tipo de mineral
proc_cierre.plot(color = 'red', ax = ax1, label = 'Proc cierre')
mantenimiento.plot(color = 'brown', ax = ax1, label = 'Mantenimiento')
produccion.plot(color = 'darkgreen', ax = ax1, label = 'Produccion')
reingenieria.plot(color = 'lime', ax = ax1, label = 'Reingenieria')
construccion.plot(color = 'darkslategrey', ax = ax1, label = 'Construccion')
factibilidad.plot(color = 'darkblue', ax = ax1, label = 'Factibilidad')
prefactibilidad.plot(color = 'cornflowerblue', ax = ax1, label = 'Prefactibilidad')
exp_avanzada.plot(color = 'purple', ax = ax1, label = 'Exp avanzada')
ev_econ_previa.plot(color = 'magenta', ax = ax1, label = 'Ev econ previa')

# agregar leyenda
ax1.legend(loc = 'lower right', title = 'Estado')
```

## 7) Mapa interactivo con Folium

Para hacer un mapa interactivo necesitaremos la librería `folium`

```python
# librerias requeridas
import folium

# crear un mapa base
m_1 = folium.Map(location = [-36.287771, -64.406169], tiles = 'openstreetmap', zoom_start = 4)

# agregar distintos tiles
folium.TileLayer('openstreetmap').add_to(m_1)
folium.TileLayer('Stamen Terrain').add_to(m_1)
folium.TileLayer('stamentoner').add_to(m_1)

# agregar los puntos correspondientes a cada yacimiento
for i in range(len(df)):
        tooltip_text = """Nombre: {}<br>
                          Mineral: {}<br>
                          Estado: {}<br>"""
        tooltip_text = tooltip_text.format(df.iloc[i]['NOMBRE'],
                                           df.iloc[i]['MINERAL_PR'],
                                           df.iloc[i]['ESTADO']
                                          )
        folium.GeoJson(df.iloc[i]['geometry'], name = df.iloc[i]['NOMBRE'], tooltip = tooltip_text).add_to(m_1)

# agregar un control de layers
folium.LayerControl().add_to(m_1)

# mostrar el mapa
m_1
```
